### Estudio de Caso:
La empresa DataTech S.A., dedicada al análisis de datos, quiere evaluar la satisfacción laboral de sus empleados sin necesidad de encuestar a todos los trabajadores. Para ello, ha decidido tomar una muestra aleatoria de 50 empleados y medir su nivel de satisfacción en una escala de 1 a 10. La media de la muestra obtenida es 7.2 y se conoce que la desviación estándar de la población es 1.5.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

### Estimar un intervalo de confianza del 95% para la media poblacional de la satisfacción laboral.

```python
# Datos de la muestra
n = 50
media_muestra = 7.2
desv_poblacion = 1.5
conf = 0.95

# Error estandar
error_e = desv_poblacion/np.sqrt(n)

# Cuartil de confianza
z_rango_confianza = norm.ppf((1+conf)/2)

# margen de error
margen_error = z_rango_confianza*error_e

# Intervalos 
intervalo_inferior = media_muestra-margen_error
intervalo_superior = media_muestra+margen_error
print(f"Intervalo de confianza del 95% para la media poblacional: {intervalo_inferior:.2f} y {intervalo_superior:.2f}")
```

```python
# Gráfico Distribución Normal
plt.figure(figsize=(9, 5))
x = np.linspace(media_muestra-4*error_e, media_muestra+4*error_e,300)
y = norm.pdf(x,media_muestra, error_e)

plt.plot(x, y, label = "Distribución normal")
plt.axvline(intervalo_inferior, ls="--", label="Limite inferior", color='orange')
plt.axvline(intervalo_superior, ls="--", label="Limite superior", color='orange')
plt.axvline(media_muestra, ls="--", label="media", color='blue')
plt.fill_between(x, 0, norm.pdf(x,media_muestra, error_e), 
                 where=((x >= intervalo_inferior) & (x <= intervalo_superior)), alpha=0.3)

plt.title("IC 95% para el grado de satisfacción laboral")
plt.xlabel("Puntos de Satisfacción")
plt.ylabel("Densidada de valores normal")
plt.legend()
plt.show()
```

### Interpretar el intervalo obtenido y explicar qué significa en términos prácticos.

El intervalo de confianza nos estaria indicando con un 95% de probabilidades de que la media poblacional de la satisfacción laboral, se encontraria en el rango de los 6.78 y 7.62 puntos, o sea que el grado de satisfacción de los trabajadores de la empresa DataTech S.A., en un rango de 10 punto, estaría en promedio dentro de este rango (6.78 y 7.62).

### Si se deseara una mayor precisión en la estimación, ¿qué podría hacer la empresa en términos de tamaño muestral?

Se deberia aumentar el tamaño de la muestra, lo que reduciria el error estandar y permitiria mayor precisión en las estimaciones.

### Si la empresa no tuviera información sobre la desviación estándar de la población, ¿qué distribución debería utilizar para el cálculo del intervalo de confianza?

Si no se tiene información de la desviación estándar de la población, la distribución elegida para inferir el intervalo de confianza tendría que ser la t de student.
